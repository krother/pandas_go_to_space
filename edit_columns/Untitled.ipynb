```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
gravity = 9.81
```

```python
times = np.arange(1500, dtype=np.int32)
```

```python
df = pd.DataFrame({'time': times})
df['gravity'] = 9.81
df['thrust'] = 0
df.loc[69:, 'thrust'] = 10.290
df['acceleration'] = df['gravity'] - df['thrust']
df['speed'] = df['acceleration'].cumsum()
df['altitude'] = 500_000 - df['speed'].cumsum()
```

```python
df['altitude'].plot()
plt.hlines(xmin=0, xmax=1500, y=0.0, color="red")
```

```python
ax = df[['altitude', 'speed']].plot()
ax.set_yscale('log')
```
