# Linear Models and Correlation

```python
import seaborn as sns

df = sns.load_dataset("penguins")
```

```python
numeric = df.iloc[:, 2:6]
numeric.head(3)
```

```python
# calculate correlation coefficients of all numeric variables
corr = numeric.corr()
sns.heatmap(data=corr, annot=True, vmin=-1.0, vmax=1.0)
```

```python
sns.scatterplot(data=df, x="body_mass_g", y="bill_depth_mm", hue="species")
```

```python
# correct way: separate linear fit by confounding factors
sns.lmplot(data=df, x="body_mass_g", y="bill_depth_mm", hue="species")
```
